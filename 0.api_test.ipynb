# API Test

The API I use is from the Clash Royale official -- https://developer.clashroyale.com/

Test in **October 2021**, the data may change when game is updated.

```python
import urllib.request
import json
```

```python
with open('api_key.txt') as f:
    my_key = f.read().rstrip('\n')
    
    base_url = 'https://api.clashroyale.com/v1'
    
    endpoint = '/cards'
    
    request = urllib.request.Request(
                    base_url+endpoint,
                    None,
                    {
                        "Authorization": "Bearer %s" % my_key
                    })
    
    response = urllib.request.urlopen(request).read().decode('utf-8')
```

**Access card information**

Get list of available cards:

```python
data = json.loads(response)

for item in data['items']:
    print('%s [%d]' % (item['name'], item['maxLevel']))
```
